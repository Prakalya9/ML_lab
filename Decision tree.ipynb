```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Create the dataset
data = {
    'SoilMoisture': [30, 45, 60, 55, 35, 70, 40, 65],
    'Temperature': [35, 30, 25, 28, 33, 22, 32, 24],
    'Rain': [0, 0, 1, 1, 0, 1, 0, 1],   # No=0, Yes=1
    'Humidity': [40, 50, 70, 65, 45, 80, 55, 75],
    'Irrigation': ['Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)

# Split features and target
X = df[['SoilMoisture', 'Temperature', 'Rain', 'Humidity']]
y = df['Irrigation']

# Train Decision Tree
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
model.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(14, 7))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=model.classes_,
    filled=True
)
plt.show()
```
