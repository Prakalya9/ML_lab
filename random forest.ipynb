```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Create the dataset
data = {
    'SoilMoisture': [30, 45, 60, 55, 35, 70, 40, 65],
    'Temperature': [35, 30, 25, 28, 33, 22, 32, 24],
    'Rain': [0, 0, 1, 1, 0, 1, 0, 1],   # No=0, Yes=1
    'Humidity': [40, 50, 70, 65, 45, 80, 55, 75],
    'Irrigation': ['Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)

# Split features and target
X = df[['SoilMoisture', 'Temperature', 'Rain', 'Humidity']]
y = df['Irrigation']

# Train Decision Tree
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
model.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(14, 7))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=model.classes_,
    filled=True
)
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Create dataset
data = {
    "Hours": [2, 4, 6, 8, 5, 3],
    "Attendance": [60, 70, 80, 90, 75, 65],
    "Pass": [0, 0, 1, 1, 1, 0]  # No=0, Yes=1
}

df = pd.DataFrame(data)

X = df[["Hours", "Attendance"]]
y = df["Pass"]

# Create Random Forest model
model = RandomForestClassifier(
    n_estimators=3,
    random_state=42
)

# Train model
model.fit(X, y)

# New student prediction
new_student = [[5, 72]]
prediction = model.predict(new_student)

print("Prediction:", "Pass" if prediction[0] == 1 else "Fail")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_text

# -----------------------------
# Step 1: Dataset
# -----------------------------
data = {
    "Hours": [2, 4, 6, 8, 5, 3],
    "Attendance": [60, 70, 80, 90, 75, 65],
    "Pass": [0, 0, 1, 1, 1, 0]  # 0 = Fail, 1 = Pass
}

df = pd.DataFrame(data)

X = df[["Hours", "Attendance"]]
y = df["Pass"]

# -----------------------------
# Step 2: Train Random Forest
# -----------------------------
rf = RandomForestClassifier(
    n_estimators=3,
    max_depth=3,
    random_state=42
)

rf.fit(X, y)

# -----------------------------
# Step 3: New data for prediction
# -----------------------------
new_student = pd.DataFrame(
    [[5, 72]],
    columns=["Hours", "Attendance"]
)

print("📌 New Student Data")
print(new_student)

# -----------------------------
# Step 4: Decision Trees + Predictions
# -----------------------------
votes = []

for i, tree in enumerate(rf.estimators_):
    print(f"\n🌳 Decision Tree {i+1}")
    print(export_text(tree, feature_names=list(X.columns)))

    tree_prediction = tree.predict(new_student)[0]
    votes.append(tree_prediction)

    print("Prediction from this tree:",
          "Pass" if tree_prediction == 1 else "Fail")

# -----------------------------
# Step 5: Final Random Forest Prediction
# -----------------------------
final_prediction = rf.predict(new_student)[0]

print("\n🗳️ Tree Votes:", ["Pass" if v == 1 else "Fail" for v in votes])
print("🌲 Random Forest Final Prediction:",
      "Pass" if final_prediction == 1 else "Fail")
```
